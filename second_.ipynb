```python
data = pd.read_csv("different.csv")
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x="Native_community", hue="readmitted", palette="Set1")
plt.xlabel("Native Community")
plt.ylabel("Count")
plt.title("Readmission by Native Community")
plt.legend(title="Readmitted", loc="upper right")
```

```python
g = sns.catplot(data=data, x="Native_community", col="readmitted", kind="count", palette="Set1")
g.set_axis_labels("Native Community", "Count")
g.set_titles("Readmission: {col_name}")
g.set_xticklabels(rotation=90)
plt.tight_layout()
```

```python
g = sns.catplot(data=data, x="gender", col="insulin", kind="count", palette="Set1")
g.set_axis_labels("gender", "Count")
g.set_titles("insulin: {col_name}")
g.set_xticklabels(rotation=90)
plt.tight_layout()
```
